```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend and format line.
plt.plot(x_axis, y_axis, label="Boston", marker="*", color="blue", linewidth=2)
plt.legend()

# Create labels for x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare By Month")

```

```python
# Create bar chart
plt.bar(x_axis,y_axis)
```

```python
# Create the plot and format the colour and add label for legend
plt.bar(x_axis, y_axis, color="green", label="Boston")

# Add legend
plt.legend()

# Create labels for x and y axis
plt.xlabel("Months")
plt.ylabel("Fare($)")

# Create a title
plt.title("Pyber Fare by Month")

```

```python
# Create the plot (horizontal bar chart)
plt.barh(x_axis, y_axis)
```

```python
# Change colors of chart
plt.barh(x_axis, y_axis, color="magenta", label="Boston")

# Add legend
plt.legend()

# Add title
plt.title("PyBer Fare by Month")

# Label x-axis
plt.xlabel("Months")

# Label y-axis
plt.ylabel("Fare($)")

# Invert y-axos so that January is at the top
plt.gca().invert_yaxis()
```

```python
# Create scatter plot 
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create scatter plot using the plt.scatter() method and change colors of markers to red
plt.scatter(y_axis, x_axis, c='red', label='Chicago')

# Add legend
plt.legend()

# Add title
plt.title("PyBer Fare by Month")

# Label x-axis
plt.xlabel("Fare ($)")

# Label y-axis
plt.ylabel("Months")

# Add limit for the x-asis data
plt.xlim(0, 50)

# Invert y-axos so that January is at the top
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Making the bubble bigger
y_axis_larger = [] # Make new list 
for data in y_axis: # Loop through y axis
  y_axis_larger.append(data*3) # multiply data in y axis by 3 and add to new list

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Refactored code for making bubble bigger
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c="skyblue", edgecolors="black", linewidths=2, s = [i * 5 for i in y_axis], label="Boston")
plt.gca().invert_yaxis()
plt.title("PyBer Fare By Month")
plt.legend()
```

```python
# Create pie chart
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Format pie chart; increase chart size, explode July, add label, add percentages
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
